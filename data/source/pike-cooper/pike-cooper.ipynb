```python
import json
import pandas
import pathlib
import uuid

dataframe = pandas.read_csv(pathlib.Path.cwd() / 'dataset.csv')

work_auth = dataframe.copy()
work_auth = work_auth[['work_uuid']].drop_duplicates()
work_auth['work_authority'] = [str(uuid.uuid4()) for x in range(len(work_auth))]
dataframe = pandas.merge(dataframe, work_auth, on='work_uuid', how='left')

agent_auth = dataframe.copy()
agent_auth = agent_auth[['agent_uuid']].drop_duplicates()
agent_auth['agent_authority'] = [str(uuid.uuid4()) for x in range(len(agent_auth))]
dataframe = pandas.merge(dataframe, agent_auth, on='agent_uuid', how='left')

for x in dataframe.work_uuid.unique():
    section = dataframe.loc[dataframe.work_uuid.isin([x])]
    name = list(section.title.unique())[0]
    gen_uuid = list(section.work_authority.unique())[0]
    data = {'uuid':gen_uuid, 'type':'work', 'name':name, 'pike-cooper':x, 'agents':list(section.agent_authority.unique())}

    path = pathlib.Path.cwd().parents[1] / 'authority' / gen_uuid[:2] / f'{gen_uuid}.json'
    path.parents[0].mkdir(parents=True, exist_ok=True)
    with open(path, 'w') as export:
        json.dump(data, export, indent=4, ensure_ascii=False)

for x in dataframe.agent_uuid.unique():
    section = dataframe.loc[dataframe.agent_uuid.isin([x])]
    name = list(section.director.unique())[0]
    gen_uuid = list(section.agent_authority.unique())[0]
    data = {'uuid':gen_uuid, 'type':'agent', 'name':name, 'pike-cooper':x, 'works':list(section.work_authority.unique())}

    path = pathlib.Path.cwd().parents[1] / 'authority' / gen_uuid[:2] / f'{gen_uuid}.json'
    path.parents[0].mkdir(parents=True, exist_ok=True)
    with open(path, 'w') as export:
        json.dump(data, export, indent=4, ensure_ascii=False)

print(len(dataframe))
dataframe.head()
```
